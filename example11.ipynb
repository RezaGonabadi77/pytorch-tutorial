```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor,Lambda
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
```

```python
train_ds = datasets.MNIST(root='dataMNIST',
                          train= True,
                          download= True,
                          transform= ToTensor(),
                          target_transform= Lambda(lambda y:torch.zeros(10).
                                                   scatter(dim = 0, index= torch.tensor(y),value = 1).type(torch.float32))
                          )
test_ds = datasets.MNIST(root='dataMNIST',
                         train=False,
                         download=True,
                         transform= ToTensor(),
                         target_transform=Lambda(lambda y:torch.zeros(10).
                                                   scatter(dim = 0, index= torch.tensor(y),value = 1).type(torch.float32))
                          )
```

```python
dl = DataLoader(train_ds,batch_size=31,shuffle= True, drop_last= False)
```

```python
for x_batch, y_batch in dl:
    print(f'train size : {x_batch.size()}')
```
