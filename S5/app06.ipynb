```python
import torch
from matplotlib import pyplot as plt
from sklearn import datasets
```

```python
data , targets = datasets.make_regression(n_samples=200, n_features=2
                                          ,n_targets=1,random_state=0, shuffle=True
                                          ,noise=2)
```

```python
plt.plot(data[:,0], targets, '.b')
plt.plot(data[:,1], targets, '.r')
plt.show()
```
