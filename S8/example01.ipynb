```python
import torch
from matplotlib import pyplot as plt
```

```python
#Sign Activation
x = torch.arange(-10,10,0.01)
y = torch.sign(x)

plt.plot(x,y,'r')
plt.grid('minor')
plt.xlabel('X', fontsize = 20)
plt.ylabel('Y', fontsize = 20)
```

```python
#Sigmoid Activation
x = torch.arange(-10,10,0.01)
y = torch.sigmoid(x)

plt.plot(x,y,'r')
plt.grid('minor')
plt.xlabel('X', fontsize = 20)
plt.ylabel('Y', fontsize = 20)
```

```python
#tanh Activation
x = torch.arange(-10,10,0.01)
y = torch.tanh(x)

plt.plot(x,y,'r')
plt.grid('minor')
plt.xlabel('X', fontsize = 20)
plt.ylabel('Y', fontsize = 20)
```

```python
#ReLU Activation
x = torch.arange(-10,10,0.1)
y = torch.relu(x)

plt.plot(x,y,'r')
plt.grid('minor')
plt.xlabel('X', fontsize = 20)
plt.ylabel('Y', fontsize = 20)
```
